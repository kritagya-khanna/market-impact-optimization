```python
# Cell 2: Data Loading Function
def load_stock_data(stock_name, data_path):
    folder_path = data_path / stock_name  
    print(f"Looking in folder: {folder_path}")
    if not folder_path.exists():
        print(f"Folder {folder_path} does not exist!")
        return None, []

    csv_files = list(folder_path.glob('*.csv'))
    print(f"Found {len(csv_files)} CSV files for {stock_name}")

    if not csv_files:
        print(f"No CSV files found for {stock_name} in {folder_path}")
        return None, []

    sample_file = csv_files[0]
    print(f"Loading sample file: {sample_file.name}")
    
    try:
        sample_data = pd.read_csv(sample_file)
        print(f"Successfully loaded {sample_data.shape[0]} rows, {sample_data.shape[1]} columns")
        return sample_data, csv_files
    except Exception as e:
        print(f"Error loading {sample_file}: {e}")
        return None, csv_files

```

```python
# Cell 10: Advanced Market Impact Calculation with Order Book Data
def calculate_market_impact_advanced(data, stock_name):
    """
    Calculate market impact using order book data with multiple levels
    """
    print(f"Calculating advanced market impact for {stock_name}")
    
    # Method 1: Best Bid/Ask Spread Analysis
    if 'bid_px_00' in data.columns and 'ask_px_00' in data.columns:
        # Best bid and ask prices
        data['best_bid'] = data['bid_px_00']
        data['best_ask'] = data['ask_px_00']
        data['mid_price'] = (data['best_bid'] + data['best_ask']) / 2
        data['spread'] = data['best_ask'] - data['best_bid']
        data['relative_spread'] = data['spread'] / data['mid_price']
        
        print(f"âœ“ Basic spread metrics calculated")
    
    # Method 2: Order Book Depth Analysis
    # Calculate total volume at each side
    bid_size_cols = [col for col in data.columns if col.startswith('bid_sz_')]
    ask_size_cols = [col for col in data.columns if col.startswith('ask_sz_')]
    
    if bid_size_cols and ask_size_cols:
        # Total order book depth
        data['total_bid_size'] = data[bid_size_cols].sum(axis=1)
        data['total_ask_size'] = data[ask_size_cols].sum(axis=1)
        data['total_book_size'] = data['total_bid_size'] + data['total_ask_size']
        
        # Order book imbalance
        data['book_imbalance'] = (data['total_bid_size'] - data['total_ask_size']) / data['total_book_size']
        
        print(f"âœ“ Order book depth calculated")
    
    # Method 3: Multi-Level Price Impact
    # Calculate weighted average prices for different order sizes
    def calculate_vwap_impact(row, side='buy', order_size=1000):
        """Calculate Volume Weighted Average Price for a given order size"""
        if side == 'buy':
            # For buy orders, we hit the ask side
            price_cols = [col for col in data.columns if col.startswith('ask_px_')]
            size_cols = [col for col in data.columns if col.startswith('ask_sz_')]
        else:
            # For sell orders, we hit the bid side  
            price_cols = [col for col in data.columns if col.startswith('bid_px_')]
            size_cols = [col for col in data.columns if col.startswith('bid_sz_')]
        
        remaining_size = order_size
        total_cost = 0
        total_filled = 0
        
        for i in range(len(price_cols)):
            if remaining_size <= 0:
                break
                
            price = row[price_cols[i]]
            size = row[size_cols[i]]
            
            if pd.isna(price) or pd.isna(size) or size <= 0:
                continue
                
            fill_size = min(remaining_size, size)
            total_cost += fill_size * price
            total_filled += fill_size
            remaining_size -= fill_size
        
        if total_filled > 0:
            return total_cost / total_filled
        else:
            return np.nan
    
    # Calculate VWAP for different order sizes
    order_sizes = [1,2, 5, 7, 10, 50, 100, 250, 500, 750, 1000, 1500, 2000, 3000, 5000, 7000, 10000]  # Different order sizes to test
    
    for size in order_sizes:
        data[f'vwap_buy_{size}'] = data.apply(
            lambda row: calculate_vwap_impact(row, 'buy', size), axis=1
        )
        data[f'vwap_sell_{size}'] = data.apply(
            lambda row: calculate_vwap_impact(row, 'sell', size), axis=1
        )
        
        # Calculate impact as difference from mid price
        data[f'impact_buy_{size}'] = (data[f'vwap_buy_{size}'] - data['mid_price']) / data['mid_price']
        data[f'impact_sell_{size}'] = (data['mid_price'] - data[f'vwap_sell_{size}']) / data['mid_price']
    
    print(f"âœ“ Multi-level VWAP impact calculated for order sizes: {order_sizes}")
    
    return data

# Apply advanced calculation to all stocks
market_impact_data = {}
for stock, data in all_stock_data.items():
    market_impact_data[stock] = calculate_market_impact_advanced(data.copy(), stock)
```

```python

def plot_market_impact_by_size(data, stock_name):
    """Plot market impact for different order sizes"""
    
    order_sizes = [1,2, 5, 7, 10, 50, 100, 250, 500, 750, 1000, 1500, 2000, 3000, 5000, 7000, 10000] 
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    for size in order_sizes:
        col_name = f'impact_buy_{size}'
        if col_name in data.columns:
            clean_data = data[data[col_name].notna() & np.isfinite(data[col_name])]
            ax1.scatter(range(len(clean_data)), clean_data[col_name], 
                       label=f'Size {size}', alpha=0.6, s=20)
    
    ax1.set_title(f'{stock_name} - Buy Order Market Impact')
    ax1.set_xlabel('Time (observation)')
    ax1.set_ylabel('Market Impact (%)')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    for size in order_sizes:
        col_name = f'impact_sell_{size}'
        if col_name in data.columns:
            clean_data = data[data[col_name].notna() & np.isfinite(data[col_name])]
            ax2.scatter(range(len(clean_data)), clean_data[col_name], 
                       label=f'Size {size}', alpha=0.6, s=20)
    
    ax2.set_title(f'{stock_name} - Sell Order Market Impact')
    ax2.set_xlabel('Time (observation)')
    ax2.set_ylabel('Market Impact (%)')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

for stock, data in market_impact_data.items():
    plot_market_impact_by_size(data, stock)
```

```python
# Cell 12: Order Size vs Market Impact Analysis
def analyze_size_impact_relationship(data, stock_name):
    
    order_sizes = [1,2, 5, 7, 10, 50, 100, 250, 500, 750, 1000, 1500, 2000, 3000, 5000, 7000, 10000] 

    impact_summary = []
    
    for size in order_sizes:
        buy_col = f'impact_buy_{size}'
        sell_col = f'impact_sell_{size}'
        
        if buy_col in data.columns and sell_col in data.columns:
            buy_impact = data[buy_col].dropna()
            sell_impact = data[sell_col].dropna()
            
            impact_summary.append({
                'order_size': size,
                'buy_impact_mean': buy_impact.mean(),
                'buy_impact_std': buy_impact.std(),
                'sell_impact_mean': sell_impact.mean(), 
                'sell_impact_std': sell_impact.std(),
                'combined_impact': (buy_impact.mean() + sell_impact.mean()) / 2
            })
    
    summary_df = pd.DataFrame(impact_summary)
    
    display(summary_df)
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(summary_df['order_size'], summary_df['buy_impact_mean'], 'bo-', label='Buy Impact')
    plt.plot(summary_df['order_size'], summary_df['sell_impact_mean'], 'ro-', label='Sell Impact')
    plt.xlabel('Order Size')
    plt.ylabel('Average Market Impact')
    plt.title(f'{stock_name} - Impact vs Order Size')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    plt.plot(summary_df['order_size'], summary_df['combined_impact'], 'go-', label='Combined Impact')
    plt.xlabel('Order Size')
    plt.ylabel('Combined Market Impact')
    plt.title(f'{stock_name} - Combined Impact vs Order Size')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return summary_df

# Analyze all stocks
impact_relationships = {}
for stock, data in market_impact_data.items():
    impact_relationships[stock] = analyze_size_impact_relationship(data, stock)
```

```python

model_ready_data = {}

for stock in stocks:
    if stock in impact_relationships:
        relationship_data = impact_relationships[stock]
        
        order_sizes = relationship_data['order_size'].values
        impacts = relationship_data['combined_impact'].values
        
        valid_idx = np.isfinite(order_sizes) & np.isfinite(impacts) & (impacts > 0)
        
        model_ready_data[stock] = {
            'order_size': order_sizes[valid_idx],
            'market_impact': impacts[valid_idx],
            'count': len(order_sizes[valid_idx])
        }
        
        print(f"{stock}: {len(order_sizes[valid_idx])} data points ready for modeling")
        print(f"  Order size range: {order_sizes[valid_idx].min():.0f} - {order_sizes[valid_idx].max():.0f}")
        print(f"  Impact range: {impacts[valid_idx].min():.6f} - {impacts[valid_idx].max():.6f}")
```

```python
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score, mean_squared_error
import warnings
warnings.filterwarnings('ignore')


def linear_model(x, beta):
    return beta * x

def sqrt_model(x, alpha):
    return alpha * np.sqrt(x)

def power_model(x, alpha, beta):
    return alpha * (x ** beta)
```

```python
def fit_models_to_stock(stock_name, order_sizes, market_impacts):
    
    print(f"\nfiting model for {stock_name}")
    
    results = {}
    
    try:
        popt_linear, _ = curve_fit(linear_model, order_sizes, market_impacts)
        predictions_linear = linear_model(order_sizes, *popt_linear)
        
        results['Linear'] = {
            'parameters': {'beta': popt_linear[0]},
            'predictions': predictions_linear,
            'r2': r2_score(market_impacts, predictions_linear),
            'rmse': np.sqrt(mean_squared_error(market_impacts, predictions_linear)),
            'aic': len(order_sizes) * np.log(np.sum((market_impacts - predictions_linear)**2) / len(order_sizes)) + 2 * 1
        }
        print(f"âœ“ Linear Model: Î² = {popt_linear[0]:.8f}")
        
    except Exception as e:
        print(f"âœ— Linear Model failed: {e}")
        results['Linear'] = None

    try:
        popt_sqrt, _ = curve_fit(sqrt_model, order_sizes, market_impacts)
        predictions_sqrt = sqrt_model(order_sizes, *popt_sqrt)
        
        results['Square Root'] = {
            'parameters': {'alpha': popt_sqrt[0]},
            'predictions': predictions_sqrt,
            'r2': r2_score(market_impacts, predictions_sqrt),
            'rmse': np.sqrt(mean_squared_error(market_impacts, predictions_sqrt)),
            'aic': len(order_sizes) * np.log(np.sum((market_impacts - predictions_sqrt)**2) / len(order_sizes)) + 2 * 1
        }
        print(f"âœ“ Square Root Model: Î± = {popt_sqrt[0]:.8f}")
        
    except Exception as e:
        print(f"âœ— Square Root Model failed: {e}")
        results['Square Root'] = None

    try:
        # Initial guess for power law parameters
        popt_power, _ = curve_fit(power_model, order_sizes, market_impacts, 
                                  p0=[0.001, 0.5], maxfev=10000)
        predictions_power = power_model(order_sizes, *popt_power)
        
        results['Power Law'] = {
            'parameters': {'alpha': popt_power[0], 'beta': popt_power[1]},
            'predictions': predictions_power,
            'r2': r2_score(market_impacts, predictions_power),
            'rmse': np.sqrt(mean_squared_error(market_impacts, predictions_power)),
            'aic': len(order_sizes) * np.log(np.sum((market_impacts - predictions_power)**2) / len(order_sizes)) + 2 * 2
        }
        print(f"âœ“ Power Law Model: Î± = {popt_power[0]:.8f}, Î² = {popt_power[1]:.4f}")
        
    except Exception as e:
        print(f"âœ— Power Law Model failed: {e}")
        results['Power Law'] = None
    
    return results

all_model_results = {}
for stock, data in model_ready_data.items():
    all_model_results[stock] = fit_models_to_stock(
        stock, 
        data['order_size'], 
        data['market_impact']
    )
```

```python
def create_performance_summary():
    performance_data = []
    
    for stock, models in all_model_results.items():
        for model_name, results in models.items():
            if results is not None:
                performance_data.append({
                    'Stock': stock,
                    'Model': model_name,
                    'RÂ²': results['r2'],
                    'RMSE': results['rmse'],
                    'AIC': results['aic'],
                    'Parameters': str(results['parameters'])
                })
    
    performance_df = pd.DataFrame(performance_data)
    return performance_df

print("\nmodel peroformance summary")
performance_summary = create_performance_summary()
display(performance_summary)

print("\naverage performance by model")
avg_performance = performance_summary.groupby('Model')[['RÂ²', 'RMSE', 'AIC']].mean()
display(avg_performance)
```

```python
def plot_model_fits(stock_name, data, results):
    
    order_sizes = data['order_size']
    market_impacts = data['market_impact']
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    fig.suptitle(f'{stock_name} - Model Fits Comparison', fontsize=16)
    
    ax1 = axes[0, 0]
    ax1.scatter(order_sizes, market_impacts, alpha=0.6, s=30, label='Actual Data', color='black')
    
    colors = ['red', 'blue', 'green']
    for i, (model_name, model_results) in enumerate(results.items()):
        if model_results is not None:
            ax1.plot(order_sizes, model_results['predictions'], 
                    color=colors[i], linewidth=2, label=f"{model_name} (RÂ²={model_results['r2']:.3f})")
    
    ax1.set_xlabel('Order Size')
    ax1.set_ylabel('Market Impact')
    ax1.set_title('All Models Comparison')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    model_names = list(results.keys())
    positions = [(0,1), (1,0), (1,1)]
    
    for i, (pos, model_name) in enumerate(zip(positions, model_names)):
        ax = axes[pos]
        model_results = results[model_name]
        
        if model_results is not None:
            ax.scatter(order_sizes, market_impacts, alpha=0.6, s=30, color='black', label='Actual')
            ax.plot(order_sizes, model_results['predictions'], 
                   color=colors[i], linewidth=2, label='Fitted')
            ax.set_title(f"{model_name} Model\nRÂ² = {model_results['r2']:.4f}, RMSE = {model_results['rmse']:.6f}")
        else:
            ax.text(0.5, 0.5, f'{model_name}\nFitting Failed', 
                   ha='center', va='center', transform=ax.transAxes)
        
        ax.set_xlabel('Order Size')
        ax.set_ylabel('Market Impact')
        ax.legend()
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

for stock, data in model_ready_data.items():
    if stock in all_model_results:
        plot_model_fits(stock, data, all_model_results[stock])
```
